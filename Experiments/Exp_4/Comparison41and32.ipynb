```python
import csv
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.ticker import NullFormatter  , ScalarFormatter
import matplotlib.axes as ax
import numpy as np
import statistics as st


threads = 2
num_client_machines = 3
connections_per_thread = [1, 8, 16, 22, 28, 32, 40, 50, 60, 100, 300, 500]
clients = []

for c in connections_per_thread:
    clients.append(c*threads*num_client_machines)

```

```python
def conv(array_string):
    new = np.zeros((len(array_string)))
    for i in range(0,len(array_string),1):
        new[i] = float(array_string[i])
    return new    
```

## Comparison between the two Experiments

```python
#USE clients
fig, ax_lst = plt.subplots(1, 1) 
plt.xlabel('Number of clients')
plt.ylabel('Throughput (ops/sec)')
plt.title("Comparison of the write-only workload measured in the two experiments ")
plt.grid('true', axis='both', fillstyle='left', linestyle='dashed',clip_on='true')
line_c1, = plt.plot(clients, WO41allClients_8_threads_throughput_means, color='blue', linestyle='solid', marker='o',markerfacecolor='blue', markersize=2)
line_c2, = plt.plot(clients, WO32allClients_8_threads_throughput_means, color='blue', linestyle='dashed', marker='o', markerfacecolor='blue',markersize=2 )
ax_lst.errorbar(clients,WO41allClients_8_threads_throughput_means, yerr=WO41allClients_8_threads_throughput_std, fmt=None, ecolor='blue', capthick=1, capsize=4)
ax_lst.errorbar(clients,WO32allClients_8_threads_throughput_means, yerr=WO32allClients_8_threads_throughput_std, fmt=None, ecolor='blue', capthick=1, capsize=4)

line_c3, = plt.plot(clients, WO41allClients_16_threads_throughput_means, color='red', linestyle='solid', marker='o',markerfacecolor='red', markersize=2)
line_c4, = plt.plot(clients, WO32allClients_16_threads_throughput_means, color='red', linestyle='dashed', marker='o', markerfacecolor='red',markersize=2 )
ax_lst.errorbar(clients,WO41allClients_16_threads_throughput_means, yerr=WO41allClients_16_threads_throughput_std, fmt=None, ecolor='red', capthick=1, capsize=4)
ax_lst.errorbar(clients,WO32allClients_16_threads_throughput_means, yerr=WO32allClients_16_threads_throughput_std, fmt=None, ecolor='red', capthick=1, capsize=4)

line_c5, = plt.plot(clients, WO41allClients_32_threads_throughput_means, color='green', linestyle='solid', marker='o',markerfacecolor='green', markersize=2)
line_c6, = plt.plot(clients, WO32allClients_32_threads_throughput_means, color='green', linestyle='dashed', marker='o', markerfacecolor='green',markersize=2 )
ax_lst.errorbar(clients,WO41allClients_32_threads_throughput_means, yerr=WO41allClients_32_threads_throughput_std, fmt=None, ecolor='green', capthick=1, capsize=4)
ax_lst.errorbar(clients,WO32allClients_32_threads_throughput_means, yerr=WO32allClients_32_threads_throughput_std, fmt=None, ecolor='green', capthick=1, capsize=4)

line_c7, = plt.plot(clients, WO41allClients_64_threads_throughput_means, color='yellow', linestyle='solid', marker='o',markerfacecolor='yellow', markersize=2)
line_c8, = plt.plot(clients, WO32allClients_64_threads_throughput_means, color='yellow', linestyle='dashed', marker='o', markerfacecolor='yellow',markersize=2 )
ax_lst.errorbar(clients,WO41allClients_64_threads_throughput_means, yerr=WO41allClients_64_threads_throughput_std, fmt=None, ecolor='yellow', capthick=1, capsize=4)
ax_lst.errorbar(clients,WO32allClients_64_threads_throughput_means, yerr=WO32allClients_64_threads_throughput_std, fmt=None, ecolor='yellow', capthick=1, capsize=4)

#ax_lst.set_xlim(0, 300)


plt.legend([ line_c1, line_c2, line_c3, line_c4, line_c5, line_c6, line_c7, line_c8 ], ['4.1 8threads','3.2 8threads', 
                                                                                       '4.1 16threads','3.2 16threads',
                                                                                       '4.1 32threads','3.2 32threads',
                                                                                       '4.1 64threads','3.2 64threads'])
plt.savefig("Write_Only_ThroughputForBothExperiments41and32.png", bbox_inches='tight')
```
